```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
plt.rcParams['font.sans-serif']='SimHei'
plt.rcParams['axes.unicode_minus']=False
```

```python
dataset=pd.read_csv('3-8.csv',encoding='gbk')
funv=dataset.groupby('order_date')['ord_qty'].sum()
ord1=funv.tolist()
day1=funv.index.tolist()
print(ord1)
print(day1)
ord=[ 40216, 35566, 51190, 61976, 56126,65090, 53548]
day=['2017-3-5', '2017-3-6', '2017-3-7', '2017-3-8', '2017-3-9','2017-3-10', '2017-3-11']
```

```python
plt.plot(day, ord, 'o-', alpha=0.5, linewidth=1, label='女神节')
plt.legend() 
plt.xlabel('date') 
plt.ylabel('order_quantity')
```
