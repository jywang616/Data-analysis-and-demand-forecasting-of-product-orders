```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

plt.rcParams['font.sans-serif']='SimHei'
plt.rcParams['axes.unicode_minus']=False

dataset=pd.read_csv('yuandan.csv',encoding='gbk')
dataset.head()
```

```python
yuandan=dataset.groupby('order_date')['ord_qty'].sum()
yuandan
```

```python
ord=yuandan.tolist()
day=yuandan.index.tolist()
plt.plot(day, ord, 'o-', alpha=0.5, linewidth=1, label='元旦')
plt.legend() 
plt.xlabel('date') 
plt.ylabel('order_quantity')
```
