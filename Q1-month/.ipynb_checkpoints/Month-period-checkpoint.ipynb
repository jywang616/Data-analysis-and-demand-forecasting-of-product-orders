```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

plt.rcParams['font.sans-serif']='SimHei'
plt.rcParams['axes.unicode_minus']=False
```

```python
dataset=pd.read_csv('2016-09.csv',encoding='gbk')
month=dataset.groupby('order_date')['ord_qty'].sum()
ord16=month.tolist()
day16=month.index.tolist()
```

```python
import matplotlib.ticker as ticker
dataset=pd.read_csv('2016-09.csv',encoding='gbk')
plt.figure(
    figsize = (20,8),
    dpi = 100        
)

# 长型数据多折线图
ax = sns.lineplot(
    x=day16,
    y=ord16,
    palette="pastel"
)
ax.xaxis.set_major_locator(ticker.MultipleLocator(base=3)) # 解决X轴密集问题
plt.show()
```
