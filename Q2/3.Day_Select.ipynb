# 处理日结果

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from datetime import datetime
from datetime import timedelta
from tqdm import tqdm
sns.set()
```

```python
out=pd.read_csv("predict_sku1.csv",encoding='utf-8')
out.head()
```

```python
outcomes=[]
i=0
for index,row in out.iterrows():
    rcode=row['sales_region_code']
    icode=row['item_code']

    name='C:/Users/Huilin Yao/Desktop/泰迪杯2023/日结果/'+'regioncode为'+str(rcode)+'item_code为'+str(icode)+'的日预测量'+'.csv'
    if os.path.exists(name) is False:
        value=[rcode,icode,0,0,0]
        outcomes.append(value)
        continue

    data=pd.read_csv(name,encoding='utf-8')
    data['日期']=pd.to_datetime(data['日期'])
    data1=data[data['日期'].dt.month.isin(np.arange(1,2))]
    data2=data[data['日期'].dt.month.isin(np.arange(2,3))]
    data3=data[data['日期'].dt.month.isin(np.arange(3,4))]
    m1=data1['预测需求量'].sum()
    m2=data2['预测需求量'].sum()
    m3=data3['预测需求量'].sum()
    value=[rcode,icode,m1,m2,m3]
    outcomes.append(value)

    i=i+1
    print('已完成 %d 数据'%i)
    
outcomes=pd.DataFrame(outcomes)
name='C:/Users/Huilin Yao/Desktop/泰迪杯2023/'+'日时间粒度的预测结果'+'.csv'
outcomes.to_csv(name,index=False)
```
